```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import mglearn

import pandas as pd

import numpy as np

%matplotlib inline
```

```python
cancer = load_breast_cancer()
print(cancer.DESCR)
```

```python
print('Cancer Keys: ' + str(cancer.keys()))
print('Cancer Feature: ' + str(cancer.feature_names))
print('Cancer Data: ' + str(cancer.data))
```

```python
df = pd.DataFrame(np.c_[cancer['data'], cancer['target']],
                  columns= np.append(cancer['feature_names'], ['target']))
df.describe().T
```

```python
print(cancer.target_names)

print("Target counts per class: {}".format(
        {n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}))

print(df['target'].value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    stratify=cancer.target, random_state=42)
print('X_train Shape: ' + str(X_train.shape))
print('y_train Shape: ' + str(y_train.shape))
print('X_test Shape: ' + str(X_test.shape))
print('y_test Shape: ' + str(y_test.shape))

#For example, if variable y is a binary categorical variable with values 0 and 1 and there are 25% of zeros and 75% of ones, stratify=y will make sure that your random split has 25% of 0's and 75% of 1's.
```

```python
#classifier = KNeighborsClassifier(n_neighbors = 6, metric = 'minkowski', p = 2)

classifier = KNeighborsClassifier(n_neighbors = 6)
classifier.fit(X_train, y_train)

print('Accuracy of KNN, on the training set: {:.3f}'.format(classifier.score(X_train, y_train)))
print('Accuracy of KNN, on the test set: {:.3f}'.format(classifier.score(X_test, y_test)))
```

```python
print("Test set predictions: {}".format(classifier.predict(X_test)))
```

```python
from sklearn.preprocessing import MinMaxScaler

mxScalar = MinMaxScaler()
X_train = mxScalar.fit_transform(X_train)
X_test = mxScalar.transform(X_test)

classifier.fit(X_train, y_train)

print('MinMaxScaler Accuracy of KNN n-6, on the training set: {:.3f}'.format(classifier.score(X_train, y_train)))
print('MinMaxScaler Accuracy of KNN n-6, on the test set: {:.3f}'.format(classifier.score(X_test, y_test)))
```

```python
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)

# Create two lists for training and test accuracies
training_accuracy = []
test_accuracy = []

# Define a range of 1 to 10 (included) neighbors to be tested
neighbors_settings = range(1,11)

# Loop with the KNN through the different number of neighbors to determine the most appropriate (best)
for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
    y_pred = clf.predict(X_test)
    print("Accuracy is ", accuracy_score(y_test,y_pred)*100,"% for K-Value:",n_neighbors)

# Visualize results - to help with deciding which n_neigbors yields the best results (n_neighbors=6, in this case)
plt.plot(neighbors_settings, training_accuracy, label='Accuracy of the training set')
plt.plot(neighbors_settings, test_accuracy, label='Accuracy of the test set')
plt.ylabel('Accuracy')
plt.xlabel('Number of Neighbors')
plt.legend()
```
